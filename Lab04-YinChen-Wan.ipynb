# Fetching the data 
Most of the data such as the chromosome which the gene is on is fecthed using the tool below. Putting out how do I get every single gene in separate cells will be too long and it lags my notebook(tried doing it in another notebook, loading the notebook itself and and running the cells take way too long). Thus I only put what I did for the Hexokinase gene below. Just changing what is the data I want in the term= argument changes the information fected.

```python
Entrez.email = 'yinchen202009@berkeley.edu'
handle = Entrez.esearch(db= 'nucleotide', 
                        term = 'Homo sapiens[ORGN] hexokinase', 
                       sort = 'relevance',
                       idtype = 'acc')

for i in Entrez.read(handle)['IdList']:
    handle = Entrez.efetch(db = 'nucleotide', id = i, rettype = 'gb', retmode= 'text')
    print(handle.read())
```
