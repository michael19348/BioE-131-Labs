```python
#Read in the datafile ‘romeoandjuliet.txt’.
import sys
import string
string.punctuation

text = open("romeoandjuliet.txt")

#Create a dictionary
wordcount=dict()
words=[]
for line in text:
    line = line.rstrip()
    line= line.translate(line.maketrans('','',string.punctuation))
    line= line.lower()
    words += line.split()
# print(words)
for word in words:
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1
for word in wordcount.items(): print("{}\t{}".format(*word))    
```

```python
#Sort dictionary according to decreasing word count
list = sorted(wordcount.items(), key = lambda x:x[1], reverse = True)
for word,freq in list:
    print ("%-10s %d" % (word, freq))
```

```python
#Graph of word count vs rank of the word
import collections
import math
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

list2=[]
for x,y in list:
    list2.append(y)

    
top= 100
plt.bar(range(0,top), list2[0:100])
plt.xlabel('Rank of the word') 
plt.ylabel('Word Count')


```

```python
#Graph og log(word count) vs. log(rank)
logx = [0,1,2,3]

list3=[]
for x in logx:
    list3.append(math.log10(list2[10**x]))
    
plt.bar(logx, list3)
plt.xlabel('Log(Rank of the word)') 
plt.ylabel('Log(Word Count)')
```

## Interpretation of Graph of word count vs rank of the word:
The log graph looks less continuous because the rank of the words are being grouped and rounded off into the log format.
